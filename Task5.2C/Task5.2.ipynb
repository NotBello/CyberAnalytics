```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load the data
data = pd.read_csv('admission_predict.csv')

# Split the data into training and testing sets
train_data = data.head(300)
test_data = data.tail(100)
```

### Need to build a linear regression function for less code :)

```python
# Function to build and plot linear regression model
def plot_linear_regression(ax, X_train, y_train, X_test, y_test, feature, color):
    # Build linear regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Plotting training set
    ax[0].scatter(X_train, y_train, color='blue', label='Training Data')
    ax[0].plot(X_train, model.predict(X_train), color=color, linewidth=2, label='Regression Line')

    # Plotting testing set and residuals
    ax[1].scatter(X_test, y_test, color='red', label='Testing Data')
    ax[1].plot(X_test, model.predict(X_test), color=color, linestyle='dashed', linewidth=2, label='Predicted Points')

    # Plot residuals as vertical lines
    ax[1].vlines(X_test, model.predict(X_test), y_test, colors='green', linestyle='dotted', label='Residuals')

    # Set labels and title
    ax[0].set_xlabel(feature)
    ax[0].set_ylabel('Chance of Admit')
    ax[1].set_xlabel(feature)
    ax[1].set_ylabel('Chance of Admit')

    ax[0].set_title(f'Linear Regression for {feature}')
    ax[1].set_title(f'True Values and Predicted Points for {feature}')

    ax[0].legend()
    ax[1].legend()
```

```python
# Set up subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(14, 10), dpi=100)

# Linear regression for GRE Score
X_train_GRE = train_data[['GRE Score']]
y_train_GRE = train_data['Chance of Admit']
X_test_GRE = test_data[['GRE Score']]
y_test_GRE = test_data['Chance of Admit']
plot_linear_regression(ax[:, 0], X_train_GRE, y_train_GRE, X_test_GRE, y_test_GRE, 'GRE Score', 'purple')

# Linear regression for CGPA Score
X_train_CGPA = train_data[['CGPA']]
y_train_CGPA = train_data['Chance of Admit']
X_test_CGPA = test_data[['CGPA']]
y_test_CGPA = test_data['Chance of Admit']
plot_linear_regression(ax[:, 1], X_train_CGPA, y_train_CGPA, X_test_CGPA, y_test_CGPA, 'CGPA', 'orange')

# Adjust layout
plt.tight_layout()
plt.show()
```
