```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python

# Set seed for reproducibility
np.random.seed(0)

# Create dataset
x, y = make_blobs(n_samples=200, centers=[[3, 2], [6, 4], [10, 5]], cluster_std=0.9)

# Plot the created dataset
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], color='blue', edgecolor='k')
plt.title('Created Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python

# K-Means clustering
kmeans = KMeans(init='k-means++', n_clusters=3, n_init=12)
kmeans_labels = kmeans.fit_predict(x)




plt.figure(figsize=(8, 6))
for cluster in range(3):
    cluster_points = x[kmeans_labels == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], edgecolor='k', label=f'Cluster {cluster}')
    for i in range(cluster_points.shape[0]):
        plt.plot([cluster_points[i, 0], cluster_points[i, 0]], [cluster_points[i, 1], cluster_points[i, 1]], color='white')

plt.title('KMeans')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.gca().set_facecolor('black')  # Set black background
plt.show()
```

```python

# Agglomerative Hierarchical clustering
agglomerative = AgglomerativeClustering(n_clusters=3, linkage='average')
agg_labels = agglomerative.fit_predict(x)

# Plot Agglomerative Hierarchical clustering result
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=agg_labels, cmap='viridis', edgecolor='k')
plt.title('Agglomerative Hierarchical')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Calculate distance matrix for Agglomerative Clustering using the input feature matrix
distance_matrix = linkage(x, method='average', metric='euclidean')

# Display dendrogram
plt.figure(figsize=(12, 8))
dendrogram(distance_matrix, orientation='top', labels=None, distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Agglomerative Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
