```python
# Import the packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Convert the file into CSV UTF-8 format using Excel
# Read the CSV file
df = pd.read_csv('./Malicious_or_criminal_attacks_breakdown-Top_five_industry_sectors_July-Dec-2019.csv', index_col=0)
```

```python
# Define colors for the four attack types
colors = ['red', 'yellow', 'blue', 'green']
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5), dpi=100)

# Grouped bar chart
df.plot(kind='bar', ax=axes[0], color=colors, edgecolor='black')
axes[0].set_title('Type of Attack by Top Five Industry Sectors')
axes[0].set_ylabel('Number of Attacks')
axes[0].set_xlabel('')
axes[0].set_xticklabels(df.index, rotation=90)
axes[0].legend(['Cyber incident', 'Theft of paperwork or data storage device', 'Rogue employee', 'Social engineering / impersonation'],
               title='Attack Types', loc='upper right')

# Stacked bar chart
df.plot(kind='bar', stacked=True, ax=axes[1], color=colors, edgecolor='black')
axes[1].set_title('Type of Attack by Top Five Industry Sectors (Stacked)')
axes[1].set_ylabel('Number of Attacks')
axes[1].set_xlabel('')
axes[1].set_xticklabels(df.index, rotation=90)
axes[1].legend(['Cyber incident', 'Theft of paperwork or data storage device', 'Rogue employee', 'Social engineering / impersonation'],
               title='Attack Types', loc='upper right')



# Show the plots
plt.show()
```
