```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the dataframe
df = pd.DataFrame({'a': np.random.randint(0, 50, size=100)})
df['b'] = df['a'] + np.random.normal(0, 10, size=100)
df['c'] = 100 - 2 * df['a'] + np.random.normal(0, 10, size=100)
df['d'] = np.random.randint(0, 50, 100)
```

```python
# Calculate corrcoef
corrcoef_ab = np.corrcoef(df['a'], df['b'])[0, 1]
corrcoef_ac = np.corrcoef(df['a'], df['c'])[0, 1]
corrcoef_ad = np.corrcoef(df['a'], df['d'])[0, 1]
```


# Visualize data with scatter plots and np.polyfit 

```python
# Scatter plot and line for df['a'] and df['b']
plt.figure(figsize=(7, 5), dpi=100)
plt.scatter(df['a'], df['b'], label=f'corrcoef = {corrcoef_ab:.2f}')
line_coef_ab = np.polyfit(df['a'], df['b'], 1)
xx_ab = np.arange(0, 50, 0.1)
yy_ab = line_coef_ab[0] * xx_ab + line_coef_ab[1]
plt.plot(xx_ab, yy_ab, color='red', lw=2)

plt.xlabel('a')
plt.ylabel('b')
plt.legend()
plt.title('Scatter plot and Linear Fit: a vs b')
plt.show()

# Scatter plot and line for df['a'] and df['c']
plt.figure(figsize=(7, 5), dpi=100)
plt.scatter(df['a'], df['c'], label=f'corrcoef = {corrcoef_ac:.2f}')
line_coef_ac = np.polyfit(df['a'], df['c'], 1)
xx_ac = np.arange(0, 50, 0.1)
yy_ac = line_coef_ac[0] * xx_ac + line_coef_ac[1]
plt.plot(xx_ac, yy_ac, color='blue', lw=2)

plt.xlabel('a')
plt.ylabel('c')
plt.legend()
plt.title('Scatter plot and Linear Fit: a vs c')
plt.show()

# Scatter plot and line for df['a'] and df['d']
plt.figure(figsize=(7, 5), dpi=100)
plt.scatter(df['a'], df['d'], label=f'corrcoef = {corrcoef_ad:.2f}')


plt.xlabel('a')
plt.ylabel('d')
plt.legend()
plt.title('Scatter plot and Linear Fit: a vs d')
plt.show()
```
