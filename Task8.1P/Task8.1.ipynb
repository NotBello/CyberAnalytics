```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer 


import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns; sns.set()
```

```python
cancer_x, cancer_y = load_breast_cancer( return_X_y = True, as_frame = True) 
```

```python
# scaling data
scaler = StandardScaler()
scaled_x = scaler.fit_transform(cancer_x)
```

```python
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(scaled_x)
```

```python
print('The original shape :', scaled_x.shape)
print('The reduced shape :', transformed_data.shape)
print('PCA component shape :', pca.components_.shape)
print('PCA components :\n', pca.components_)
```

```python

plt.figure(figsize=(10, 8))
scatter = plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=cancer_y, cmap="PuOr")
plt.title("2D Plot with PCs")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")

legend_labels = ['Malignant', 'Benign']

plt.legend(handles = scatter.legend_elements()[0], labels=legend_labels, title="Classes")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(scaled_x[:, 0], scaled_x[:, 1], scaled_x[:, 2], c=cancer_y, cmap="Spectral")
ax.set_title("3D Plot with First 3 Features")
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_zlabel("Feature 3")

ax.view_init(elev=10, azim=10)  

legend_labels = ['Malignant', 'Benign']

plt.legend(handles = scatter.legend_elements()[0], labels=legend_labels, title="Classes")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(transformed_data[:, 0], transformed_data[:, 1], c=cancer_y, cmap="Spectral", marker='o')
ax.set_title("3D Plot with First And Second Principal Components")
ax.set_xlabel("First principal component")
ax.set_ylabel("Second principal component")
ax.set_zlabel("")



ax.legend(handles=scatter.legend_elements()[0], labels=legend_labels, title="Classes")


ax.view_init(elev=10, azim=10) 

legend_labels = ['Malignant', 'Benign']

plt.legend(handles = scatter.legend_elements()[0], labels=legend_labels, title="Classes")
plt.show()
```
