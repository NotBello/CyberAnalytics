```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the CSV file
df = pd.read_csv('./reported_cyber_crime_related_incidents_2014_2020.csv')
```

```python
# Print columns for testing
print(df.columns)
```

```python
# Update the column name if needed
df_2020 = df[['incident_type', 'no.of_ incidents_2020']]
# Set the index to incident_type
df_2020.set_index('incident_type', inplace=True)
```

```python
print(df_2020)
```

## Unordered Bar Char of 2020

```python
colors =  ['red', 'magenta', 'yellow', 'green', 'cyan', 'blue']
```

```python
fig, ax = plt.subplots(figsize=(7, 5), dpi=100)
df_2020['no.of_ incidents_2020'].plot(kind='bar', color=colors, ax=ax)

# Set labels and title
ax.set_ylabel('No. of incidents')
ax.set_xlabel('Incident type')
ax.set_title('Number of cyber-crime related incidents recorded in Sri Lanka in 2020')

# Set x-axis labels with rotation
ax.set_xticklabels(df_2020.index, rotation=90)

# Show the plot
plt.show()

```

## Ordered Bar Char for 2020

```python
df_2020_ordered = df_2020.sort_values(by='no.of_ incidents_2020')
```

```python
fig, ax = plt.subplots(figsize=(7, 5), dpi=100)
df_2020_ordered['no.of_ incidents_2020'].plot(kind='bar', color=colors, ax=ax)

# Set labels and title
ax.set_ylabel('No. of incidents')
ax.set_xlabel('Incident type')
ax.set_title('Number of cyber-crime related incidents recorded in Sri Lanka in 2020')

# Set x-axis labels with rotation
ax.set_xticklabels(df_2020_ordered.index, rotation=90)

# Show the plot
plt.show()
```

```python
# Set the index to incident_type
df.set_index('incident_type', inplace=True)

# Select the relevant columns
df_incidents = df[['no.of_ incidents_2014', 'no.of_ incidents_2020']]

# Set the colors for the incident categories
colors = ['red', 'magenta', 'yellow', 'green', 'cyan', 'blue']

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 8))

# Plot for 2014
axs[0].pie(df_incidents['no.of_ incidents_2014'], labels=df_incidents.index, autopct='%1.1f%%', colors=colors)
axs[0].set_title('Cyber crime incidents reported in Sri Lanka in 2014')

# Plot for 2020
axs[1].pie(df_incidents['no.of_ incidents_2020'], labels=df_incidents.index, autopct='%1.1f%%', colors=colors)
axs[1].set_title('Cyber crime incidents reported in Sri Lanka in 2020')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
